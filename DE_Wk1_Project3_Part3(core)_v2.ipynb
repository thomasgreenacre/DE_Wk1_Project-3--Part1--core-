```python
## create a col with a list of genres
basics_df['genres_split'] = basics_df['genres'].str.split(',')
basics_df
```

```python
exploded_genres = basics_df.explode('genres_split')
exploded_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
```

```python
#save the tconst and genres_split to a new df
title_genres = exploded_genres[['tconst', 'genres_split']].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
## make new integer genre_id and drop string genres
basics_df['genre_id'] = basics_df['genres_split'].map(genre_map)
basics_df = basics_df.drop(columns='genres_split')
```
